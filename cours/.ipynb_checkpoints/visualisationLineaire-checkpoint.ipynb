```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
data = pd.read_csv("../data/ibm_hr_attrition.csv")
data.columns
```

```python
data['Attrition'] = data.Attrition.replace(["Yes", "No"], [1, 0])
sns.lmplot(x='DistanceFromHome', y='Attrition', data=data)
#On constate que plus un employé habite loin, plus la probabilité qu'il démissionne est importante.
```

```python
sns.lmplot(x='YearsWithCurrManager', y='Attrition', data=data)
#On constate que plus un employé travaille avec le même manager, plus la probabilité qu'il démissionne est faible.
```
